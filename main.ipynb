```python
import requests

def get_stock_news(tickers, api_key):
    # ‡∏Å‡∏≥‡∏´‡∏ô‡∏î URL ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö News Sentiment API
    url = f'https://www.alphavantage.co/query?function=NEWS_SENTIMENT&tickers={tickers}&apikey={api_key}'
    
    try:
        response = requests.get(url)
        data = response.json()
        
        # ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏ß‡πà‡∏≤‡∏°‡∏µ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• feed ‡∏´‡∏£‡∏∑‡∏≠‡πÑ‡∏°‡πà
        if "feed" not in data:
            print("‡πÑ‡∏°‡πà‡∏û‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏Ç‡πà‡∏≤‡∏ß ‡∏´‡∏£‡∏∑‡∏≠ API Limit ‡∏Ñ‡∏£‡∏ö‡πÅ‡∏•‡πâ‡∏ß")
            return []
        
        news_list = data["feed"]
        return news_list

    except Exception as e:
        print(f"‡πÄ‡∏Å‡∏¥‡∏î‡∏Ç‡πâ‡∏≠‡∏ú‡∏¥‡∏î‡∏û‡∏•‡∏≤‡∏î: {e}")
        return []

# --- ‡∏ï‡∏±‡πâ‡∏á‡∏Ñ‡πà‡∏≤‡∏Å‡∏≤‡∏£‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô ---
MY_API_KEY = 'BM2HZYW4F783EEEJ'
WATCHLIST = 'AAPL,TSLA,NVDA' # ‡πÉ‡∏™‡πà‡∏ä‡∏∑‡πà‡∏≠‡∏´‡∏∏‡πâ‡∏ô‡∏ó‡∏µ‡πà‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡∏Ñ‡∏±‡πà‡∏ô‡∏î‡πâ‡∏ß‡∏¢‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡∏´‡∏°‡∏≤‡∏¢ ,

# ‡∏£‡∏±‡∏ô‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏î‡∏π‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå
news_feed = get_stock_news(WATCHLIST, MY_API_KEY)

print(f"--- ‡∏£‡∏≤‡∏¢‡∏á‡∏≤‡∏ô‡∏Ç‡πà‡∏≤‡∏ß‡∏´‡∏∏‡πâ‡∏ô‡∏•‡πà‡∏≤‡∏™‡∏∏‡∏î ({WATCHLIST}) ---\n")
for news in news_feed[:5]: # ‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏• 5 ‡∏Ç‡πà‡∏≤‡∏ß‡∏•‡πà‡∏≤‡∏™‡∏∏‡∏î
    print(f"üìå ‡∏´‡∏±‡∏ß‡∏Ç‡πâ‡∏≠: {news['title']}")
    print(f"üì∞ ‡πÅ‡∏´‡∏•‡πà‡∏á‡∏ó‡∏µ‡πà‡∏°‡∏≤: {news['source']} | ‡πÄ‡∏ß‡∏•‡∏≤: {news['time_published']}")
    print(f"üìä Sentiment: {news['overall_sentiment_label']} (Score: {news['overall_sentiment_score']})")
    print(f"üîó ‡∏•‡∏¥‡∏á‡∏Å‡πå: {news['url']}")
    print("-" * 50)
```
